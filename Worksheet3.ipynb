Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
class?

```python
#Add a column to the Titanic dataset, fare per year, containing the fare divided by the age ofthe passenger(i.e., Fare/Age).
dataframe['fare_per_class']= dataframe['Fare']/dataframe['Pclass']
# dataframe.head()

# 2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equalto 18), assigning this to adult males.
adult_males= dataframe[(dataframe["Age"]>=18) & (dataframe["Sex"]=="male")]
# print(adult_males.head())

#3. Sort adult males by descending fare per class, assigning this to adult males srt.
adult_males_srt= adult_males.sort_values(by="fare_per_class",ascending=False)
# print(adult_males_srt.head(6))

#Select only the Name, Age, and fare per class columns of adult males sr and save the results result.
result= adult_males_srt[["Name","Age","fare_per_class"]]
print(result.head(6))
```

#3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:


```python
#1. Calculate the total Fare paid across all passengers in the Titanic dataset.
total_fare= dataframe['Fare'].sum()
print(f"the total sum of fare column is {total_fare}")


#2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
pclassones = dataframe.loc[dataframe["Pclass"] == 1, "Fare"].sum()
print(f"Total Fare paid by first-class passengers: {pclassones}")

#Do the same for second class (Pclass is 2) and third class (Pclass is 3).
totalpclass2_3= dataframe.loc[dataframe["Pclass"].isin([2,3]),"Fare"].sum()
print(f"total fare paid by 2 and 3 is {totalpclass2_3}")

#4. Combine the fare totals from first, second, and third classes into a list.
# fare_total= total_fare.tolist();
# print(fare_total)

#5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.

one = pclassones/total_fare
threeandtwo = totalpclass2_3/total_fare
print(f"the proportion for class 1 is {one}")
print(f"the proportion for class 2 and class 3 is {threeandtwo}")


```

What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?

```python
dataframe['age_group'] = 'adult'   # default

dataframe.loc[dataframe['Age'] < 18, 'age_group'] = 'child'
dataframe.loc[dataframe['Age'] >= 65, 'age_group'] = 'senior'
# dataframe.head()

#2. Calculate the total number of passengers on the Titanic.
totalpassenger= dataframe["PassengerId"].count()
print(f"there are {totalpassenger} passenger in titanic")

#3. Count the number of passengers in each age group.
age_group_counts = dataframe['age_group'].value_counts()
print(age_group_counts)
print("_______________________________________________________")

#4. Divide the count of each age group by the total number of passengers to get the proportion of passengersin each age group.

proportion = age_group_counts / totalpassenger
print(f"the proportion for each group is \n {proportion}")
print("_______________________________________________________")
#5. Display the proportion as a percentage.
intopercentage= proportion*100
print(f"into percentage is {intopercentage}")
```
